```python
def print_graph(nodes: list[Node], edges: list[Edge]):
    G = nx.Graph()
    pos = {}
    labels = {}  #Erstelle ein Label-Dictionary um nur x und y koordinate darzustellen.
    for node in nodes:
        G.add_node(node)
        pos[node] = (node.getx(), node.gety())  #Entferne die Invertierung der Y Koordinate
        labels[node] = f"({node.getx()}, {node.gety()})"  #FÃ¼ge die x und y Koordinate als String ins Label ein

    for edge in edges:
        G.add_edge(edge.getnode1(), edge.getnode2())

    nx.draw(G, pos, labels=labels, node_color='lightblue', node_size=500, font_size=8,
            width=2)
    plt.show()


print_graph(graph.getnodes(), graph.getedges())
```
